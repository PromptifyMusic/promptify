```python
# Drop additional NaN columns
df_spoti_fresh = df_spoti_fresh.drop(columns=["album_image_url", "orig_track_id"])
df_spoti_fresh[df_spoti_fresh.isna().any(axis=1)]
```

Old spotify data dataset

```python
import json
```

```python
with open("tracks.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# get needed columns to dataset
rows = []
for item in data:
    rows.append({
        "spotify_id": item.get("id"),
        "track_name": item.get("name"),
        "artist_name": item.get("artists", [{}])[0].get("name", ""),
        "album_name": item.get("album", {}).get("name", ""),
        "popularity": item.get("popularity"),
        "spotify_url": item.get("external_urls", {}).get("spotify", ""),
        "explicit": item.get("explicit"),
        "album_images": json.dumps(item.get("images"), ensure_ascii=False),
    })
```

```python
# save to dataset
df_spoti_old = pd.DataFrame(rows)
df_spoti_old.sample(5)
```

Dataset of songs not found in fresh spotify dataset

```python
df_not_found = pd.read_csv("not_found.csv")
df_not_found.sample(5)
```

Create dataframe with old spotify data of tracks not found in fresh spoti data

```python
# from df_spoti_old get rows that are not found in df_spoti_fresh
df_not_found_spoti_old = df_spoti_old[df_spoti_old["spotify_id"].isin(df_not_found["spotify_id"])] \
    .merge(
        df_not_found[["spotify_id", "track_id"]],
        on="spotify_id",
        how="left"
    ).rename(columns={"track_id": "org_track_id"})

df_not_found_spoti_old.shape
```

```python
df_not_found_spoti_old.sample(5)
```

```python
df_not_found_spoti_old[df_not_found_spoti_old.isna().any(axis=1)]
```

Create dataframe with fresh and old (not found in fresh) spotify data

```python
df_spoti_merged = pd.concat([df_spoti_fresh, df_not_found_spoti_old], ignore_index=True)
df_spoti_merged.sample(5)
```

```python
df_spoti_merged[df_spoti_merged.isna().any(axis=1)]
```

```python
import numpy as np

# replace nulls in "album_images" with NaN
df_spoti_merged["album_images"] = df_spoti_merged["album_images"].replace("null", np.nan)
df_spoti_merged[df_spoti_merged.isna().any(axis=1)]
```
